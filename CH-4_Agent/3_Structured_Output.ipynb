```python
from langchain_openai import ChatOpenAI

import os
from dotenv import load_dotenv
load_dotenv()
llm = ChatOpenAI(model = "gpt-5-mini")
```

### 1. Low-Lifting Approach:

#### Low-Lifting Approach: Get the results in the way you want / in that schema you want.

We can rely on the LLM for such kind of tasks

```python
response = llm.invoke("What is the capital of France?. Generate the output with this format:: " 'Query: <the query>'" 'Answer': <the answer>' ")
response = response.content
response
```

### Heavy-Lifting Approach:

#### Heavy-Lifting Approach: Let's suppose we've well-defined schema and we cannot compromise with that. So to achieve such kind of scenario we have to rely to Heavy-Lifting Approach, which we implement it by using Pydantic(useful for Schema Validation & Schema Parsing both the things) library in Python.

##### Pydantic

```python
from pydantic import BaseModel, Field

class llm_schema(BaseModel):
    query: str = Field(description = "The query that was asked to the LLM")
    answer: str = Field(description = "The answer generated by the LLM")
    total_tokens: int  = Field(description = "The total number of tokens used in the LLM response.")
    
```

```python
obj = llm_schema(query = "What is the capital of France?", answer = "The capital of France is Paris.", total_tokens = 15)
obj
```

```python
llm_structured = llm.with_structured_output(llm_schema)
response = llm_structured.invoke("What is the capital of France?")
response
```
