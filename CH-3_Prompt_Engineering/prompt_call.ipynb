```python
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from dotenv import load_dotenv
load_dotenv()

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
```

#### 1. Zero-Shot Prompting (Prompt without giving any example)

```python
llm.invoke("Translate the following sentence to Japnese: 'How are you doing?, where do you stay exactly?'")
```

#### 2. One-Shot Prompting: atleast provide one example to model

```python
one_shot_prompt = (
    "Translate the following English sentence to Japnese.\n"
    "Example: 'Are you well?' -> 'お元気ですか？ O-genki desu ka?' \n"
    "Now, translate: How are you doing?, where do yu stay exactly?"
)

llm.invoke(one_shot_prompt)
```

#### 3. Few-Shot Prompting: To provide multiple examples for better analzation by llm and llm will provide us the results.

```python
few_shot_prompt = (
    "Translate the following English sentence to Japnese.\n"
    "Example1: 'Hi' -> 'こんにちは Konnichiwa'\n"
    "Example2: 'Are you well?' -> 'お元気ですか？ O-genki desu ka?' \n"
    "Now, translate: Hi, How are you doing?, Thankyou!, How about you?"
)

llm.invoke(few_shot_prompt)
```

#### 4. Chain-of-Thought Prompting: Whenever you work with LLMs, so mostly we will be working with Data Stuffs(write sql query and all). So LLMs model will not perform exactly and we could not rely on model, It cannot think so much steps (step-by-step approach) and give it all correctly. So we use Chain-of-prompting means related questions for every steps.

```python
cot_prompt = (
    "Solve the following math problem step by step. \n"
    "Question: If you've 3 Apples and get 2 more, how many apples do you have?\n"
    "Let's think step by step.\n"
    "Step 1: Start with 3 Apples.\n"
    "Step 2: Get 2 more Apples.\n"
    "Step 3: 3 + 2 = 5.\n"
    "Answer: 5 Apples."
)

llm.invoke(cot_prompt)
```
